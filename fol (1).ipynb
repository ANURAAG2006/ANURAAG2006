```python
import streamlit as st
import folium
from folium.plugins import HeatMap
from streamlit_folium import folium_static

# Streamlit app
st.title("Heatmap in Streamlit using Folium")

# Sample data: [latitude, longitude, intensity]
data = [
    [13.0827, 80.2707, 0.8],  # Chennai
    [12.9716, 77.5946, 0.6],  # Bangalore
    [28.7041, 77.1025, 0.7],  # Delhi
    [19.0760, 72.8777, 1.0],  # Mumbai
    [22.5726, 88.3639, 0.5],  # Kolkata
]

# Create Folium map
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Add HeatMap layer
HeatMap(data, radius=15, blur=10, max_zoom=10).add_to(m)

# Display map in Streamlit
folium_static(m)
```

```python

import pydeck as pdk

# Define data
data = [{"lat": 37.7749, "lon": -122.4194, "weight": 10}]

# Create a layer
layer = pdk.Layer(
    "HeatmapLayer",
    data,
    get_position=["lon", "lat"],
    get_weight="weight",
)

# Create a deck
view_state = pdk.ViewState(latitude=37.7749, longitude=-122.4194, zoom=10)
r = pdk.Deck(layers=[layer], initial_view_state=view_state)

r.show()
```

```python
import osmnx as ox
import folium
import geopandas as gpd

# Define the place name (Change this to your location)
place_name = "Namakkal, Tamil Nadu, India"

# Fetch administrative boundaries (wards)
gdf = ox.geocode_to_gdf(place_name)

# Extract boundary polygon
ward_boundary = gdf.geometry.iloc[0]  # First result

# Create a Folium Map
m = folium.Map(location=[ward_boundary.centroid.y, ward_boundary.centroid.x], zoom_start=10)

# Convert boundary to GeoJSON and add to Folium
folium.GeoJson(
    ward_boundary,
    name="Ward Boundary",
    style_function=lambda x: {
        "fillColor": "none",
        "color": "black",  # Darker boundary
        "weight": 3,  # Line thickness
        "opacity": 0.8
    },
).add_to(m)

# Display map
m.save("ward_map.html")
m
```

```python
import pydeck as pdk
import pandas as pd

# Define locations with coordinates and population density
data = pd.DataFrame({
    "name": ["Tiruchengode", "Rasipuram", "Namakkal", "Paramathi-Velur"],
    "population_density": [765, 416, 433, 406],  # People per sq km
    "coordinates": [
        [[77.8961, 11.3801], [77.8961, 11.4001], [77.9161, 11.4001], [77.9161, 11.3801]],  # Tiruchengode
        [[78.1665, 11.4674], [78.1665, 11.4874], [78.1865, 11.4874], [78.1865, 11.4674]],  # Rasipuram
        [[78.1652, 11.2213], [78.1652, 11.2413], [78.1852, 11.2413], [78.1852, 11.2213]],  # Namakkal
        [[78.0313, 11.2412], [78.0313, 11.2612], [78.0513, 11.2612], [78.0513, 11.2412]],  # Paramathi-Velur
    ]
})

# Define the Pydeck layer for polygons
polygon_layer = pdk.Layer(
    "PolygonLayer",
    data,
    get_polygon="coordinates",
    get_fill_color="[population_density * 0.5, 100, 255 - population_density * 0.5, 150]",  # Color based on density
    get_line_color=[255, 255, 255],
    pickable=True,
    stroked=True,
    filled=True,
    extruded=True,
    wireframe=True,
    get_elevation="population_density * 2",  # Elevation based on population density
)

# Set the viewport
view_state = pdk.ViewState(
    latitude=11.4, longitude=78.1, zoom=9, pitch=40, bearing=0
)

# Render the map
r = pdk.Deck(
    layers=[polygon_layer],
    initial_view_state=view_state,
    tooltip={"text": "{name}\nPopulation Density: {population_density} per sq km"},
)

r.show()
```

```python
import osmnx as ox
import geopandas as gpd
import pydeck as pdk
import pandas as pd

# Define places and population density
places = {
    "Tiruchengode, Tamil Nadu, India": 765,
    "Rasipuram, Tamil Nadu, India": 416,
    "Namakkal, Tamil Nadu, India": 433,
    "Paramathi Velur, Tamil Nadu, India": 406,
}

# Function to get boundary coordinates
def get_boundary_coordinates(place):
    gdf = ox.geocode_to_gdf(place)
    if not gdf.empty:
        return list(gdf.geometry.iloc[0].exterior.coords)  # Extract polygon boundary
    return None

# Prepare Pydeck data
polygon_data = []
for place, density in places.items():
    boundary = get_boundary_coordinates(place)
    if boundary:
        polygon_data.append({"place": place, "polygon": boundary, "density": density})

# Convert to DataFrame
df = pd.DataFrame(polygon_data)

# Define a Pydeck PolygonLayer
polygon_layer = pdk.Layer(
    "PolygonLayer",
    df,
    get_polygon="polygon",
    get_fill_color="[density / 4, 100, 255 - density / 4, 140]",  # Color based on density
    pickable=True,
    extruded=True,
    get_elevation="density * 2",  # Extrusion based on population density
)

# Create Pydeck Map
view_state = pdk.ViewState(latitude=11.4, longitude=78.1, zoom=9)

# Render Map
r = pdk.Deck(layers=[polygon_layer], initial_view_state=view_state, tooltip={"text": "{place}: {density} per sq km"})
r.to_html("population_density_pydeck.html")
r
```
